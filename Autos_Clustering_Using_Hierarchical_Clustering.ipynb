```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in df.columns:
  if df[column].dtype == 'object':
    df[column] = le.fit_transform(df[column])
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(16,10))
dendogram = sch.dendrogram(sch.linkage(df,method='ward'))
plt.title('Dendogram')
plt.xlabel('Autos')
plt.ylabel('Euclident Distance')
plt.show()
```

```python

from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='average')
y_hc = hc.fit_predict(df)
y_hc
```

```python
X = df.iloc[ : ,[1,2]].values
X
```

```python
plt.figure(figsize=(16,8))

plt.scatter(X[y_hc == 0,0],X[y_hc == 0,1],color='grey',label='cluster 1',s=50)
plt.scatter(X[y_hc == 1,0],X[y_hc == 1,1],color='blue',label='cluster 2',s=50)
plt.scatter(X[y_hc == 2,0],X[y_hc == 2,1],color='black',label='cluster 3',s=50)
plt.scatter(X[y_hc == 3,0],X[y_hc == 3,1],color='yellow',label='cluster 4',s=50)
plt.scatter(X[y_hc == 4,0],X[y_hc == 4,1],color='red',label='cluster 5',s=50)
plt.title('Hierarchy clustering')
plt.xlabel('aspiration')
plt.ylabel('No of Doors')
plt.show()
```
